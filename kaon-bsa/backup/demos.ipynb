```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.rc('font', size=18)
plt.rc('font', family='serif')
```

```python
x = np.arange(400,600)
```

```python
from scipy.stats import binom
```

```python
plt.figure(figsize=(8,6))
plt.bar(x, binom.pmf(x,1000,0.5), width=1, color='red', alpha=0.5)
plt.bar(x[:70], binom.pmf(x[:70],1000,0.5), 
        color='red', width=1, alpha=0.3, hatch='///', label='two sided p-value')
plt.bar(x[130:], binom.pmf(x[130:],1000,0.5), 
        color='red', width=1, alpha=0.3, hatch='///')
plt.axvline(x[130], color='black', linestyle='-.', label='obs = %.1f' % x[130])
plt.xlabel('n')
plt.ylabel('P(n,1000,0.5)')
plt.savefig('image/p_value.pdf', bbox_inches='tight')
```

```python
from scipy.stats import norm
```

```python
MU = 0.85
SIGMA = 0.02

beta = np.linspace(MU - 5*SIGMA, MU+5*SIGMA, 100)
p    = norm.pdf(beta, MU, SIGMA)

plt.plot(beta, p, linestyle='--', color='black')
plt.fill_between(beta[:30], p[:30], color='red', alpha=0.3)
plt.fill_between(beta[70:], p[70:], color='red', alpha=0.3)
plt.axvline(MU, color='black', linestyle='-.', linewidth=1, alpha=0.8)
plt.axvline(beta[30], color='black', linestyle='-.', linewidth=1, alpha=0.8)
```
