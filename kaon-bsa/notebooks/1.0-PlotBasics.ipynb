```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 

%matplotlib inline
```

```python
data = pd.read_csv('../database/phi/sys.csv')
```

```python
data.head()
```

```python
font_settings = {'size' : 18}
text_settings = {'usetex' : True}
plt.rc('font', **font_settings)
plt.rc('text', **text_settings)
```

```python
def plot_simple_asym(data, axis, bin):
    
    d = data.query('axis == "{}" and axis_bin == {}'.format(
            axis, bin)
        )
        
    plt.errorbar(
        d['phi'], d['value'], d['stat'],
        linestyle='', marker='o', color = 'k')
    
    plt.axhline(0, linestyle='--', alpha=0.8, 
        linewidth=0.5, color='k')
        
    plt.ylim([-0.08, 0.08])
    plt.xlabel('$\phi_h$ (degrees)')
    plt.ylabel('BSA')
```

```python
plt.figure(figsize=(8,6))
plot_simple_asym(data, 'x', 2)
plt.savefig('../image/basic_bsa.pdf')
```
