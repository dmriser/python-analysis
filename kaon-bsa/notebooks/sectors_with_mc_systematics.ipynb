### Exercise 6: 
Write a function that performs the above and accepts the sectors and datapoint as arguments.

```python
def scatter_two_sectors(data, sectors, index, 
                       fig=None, ax=None):
    """ 
    Create a scatterplot of the MC trials 
    for a datapoint given by index.  Return the 
    figure. 
    
    Inputs:
    -------
    data - the results array from above indexed by (sector, datapoint, trial)
    sectors - a tuple of length 2 that specifies the two sectors to compare
    index - the datapoint to plot
    
    Returns: 
    --------
    fig - The figure object for drawing.
    ax - The axis object for drawing. 
    
    """
    
    assert(len(sectors) == 2)
    
    if fig is None:
        fig = plt.figure()
        ax = fig.add_subplot(1,1,1)
    
    if ax is None:
        ax = fig.add_subplot(1,1,1)
        
    ax.scatter(data[sectors[0]-1, index, :], data[sectors[1]-1, index, :], edgecolor='k')
    ax.axvline(data[sectors[0]-1, index, :].mean(), alpha=0.6)
    ax.axhline(data[sectors[1]-1, index, :].mean(), alpha=0.6)
    ax.grid(alpha = 0.5)
    ax.set_title('Sectors {0} ({1:6.4f}) and {2} ({3:6.4f})'.format(
        sectors[0],
        data[sectors[0]-1, index, :].mean(),
        sectors[1],
        data[sectors[1]-1, index, :].mean()
    ))
    
    # Set the scale for both axes the same so that
    # if there is disagreement the y-sector will be 
    # offset. 
    axisrange = (np.mean(data[sectors[0]-1, index, :]) - 12 * np.std(data[sectors[0]-1, index, :]),
                 np.mean(data[sectors[0]-1, index, :]) + 12 * np.std(data[sectors[0]-1, index, :]))
    ax.set_xlim(axisrange)
    ax.set_ylim(axisrange)
    
    return fig, ax
```

```python
fig, ax = scatter_two_sectors(data=results, sectors=(1,2), index=15)
```

### Exercise 7: 
Choose one datapoint and compare all the sectors using the function above.  Plot the results in a grid.

```python
index = 30

fig = plt.figure(figsize=(24,24))

for isect in range(1,7):
    for jsect in range(1,7):
        iax = (jsect - 1) + 6 * (isect - 1) + 1
        ax = fig.add_subplot(6, 6, iax)
        fig, ax = scatter_two_sectors(data = results, 
                           sectors = (isect, jsect),
                           index = index, fig = fig, ax = ax)
        
fig.tight_layout()
```
