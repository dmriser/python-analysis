```python
def plot_values(values_matrix, test_point):
    
    plt.figure(figsize=(16,10))

    for s in range(6):
        plt.subplot(2,3,s+1)
        plt.plot(np.arange(nsamples), values_matrix[test_point, s, :]);
        plt.axhline(np.mean(values_matrix[test_point, s, :]), 
                   color='k', alpha=0.4)
        plt.grid(alpha=0.2)
        plt.ylim([-0.15, 0.15])
        plt.title(f'Sector {s+1}')
```

```python
def histogram_values(values_matrix, test_point):
    
    opts = {'edgecolor':'red', 'color':'yellow', 'hatch':'///',
           'bins':np.linspace(-0.2, 0.2, 40)}
    
    plt.figure(figsize=(16,10))

    for s in range(6):
        plt.subplot(2,3,s+1)
        plt.hist(values_matrix[test_point, s, :], **opts);
        plt.grid(alpha=0.2)
        plt.title(f'Sector {s+1}')
```

```python
histogram_values(values_matrix, test_point=5)
```

```python
def plot_histogram_grid(values_matrix, test_point):
    """ This creates a 6x6 grid, half of which is redundant.  It can 
    be created without the redundancy but it is just fine like this.  
    
    The function does take some time to run, be patient. """
    opts1 = {'edgecolor':'k', 'color':'red', 'alpha':0.6,
           'bins':np.linspace(-0.2, 0.2, 40)}
    opts2 = {'edgecolor':'k', 'color':'blue', 'alpha':0.6,
           'bins':np.linspace(-0.2, 0.2, 40)}
    
    plt.figure(figsize=(16,16))
    
    for s1 in range(6):
        for s2 in range(6):
            plt.subplot(6, 6, s1 * 6 + s2 + 1)
            plt.hist(values_matrix[test_point, s1, :], **opts1);
            plt.hist(values_matrix[test_point, s2, :], **opts2);
            plt.grid(alpha=0.2)
            plt.title(f'Sector {s1+1}, {s2+1}')
            
    plt.tight_layout()
```

```python
plot_histogram_grid(values_matrix, test_point=6)
```

# Compare Measurements With Mean/Std from Boostraps

```python
values = np.mean(values_matrix, axis=2)
errors = np.std(values_matrix, axis=2)
```

```python
def calculate_weighted_z_score(values, errors):
    """ Calculate a weighted chi^2. This is PDG style. """
    
    npoints, nsectors = values.shape
    assert(values.shape == errors.shape)
    
    weights = 1. / errors**2
    mu = np.sum(values * weights, axis = 1) / np.sum(weights, axis = 1)
    sigma = np.sqrt(1 / np.sum(weights, axis = 1))
    residual = (values - np.multiply(np.ones(shape=(nsectors,npoints)), mu).transpose())
    z_score = residual / np.multiply(np.ones(shape=(nsectors,npoints)), sigma).transpose()
    
    return mu, sigma, z_score
```

```python
mu, sigma, z_score = calculate_weighted_z_score(values, errors)
```

```python
chi_sq = np.sum(z_score**2, axis=1)
chi_sq.shape
```

```python
reduced_chi_sq = chi_sq / 6.
```

```python
plt.figure(figsize=(12,5))
plt.bar(np.arange(len(reduced_chi_sq)), reduced_chi_sq)
plt.grid(alpha=0.2)
```

```python
sector_chi_sq = np.sum(z_score**2, axis=0) / 240.
```

```python
plt.figure(figsize=(12,5))
plt.bar(np.arange(len(sector_chi_sq)), sector_chi_sq)
plt.grid(alpha=0.2)
```

# Compare That to Normal Thing

```python
one_pass_data = {}
op_values = np.empty(shape=(240,6))
op_errors = np.empty(shape=(240,6))

for i in range(1,7):
    one_pass_data[i] = pd.read_csv(data_dir + f'/../sector_{i}.csv')
    op_values[:, i - 1] = one_pass_data[i]['value'].values    
    op_errors[:, i - 1] = one_pass_data[i]['stat'].values
```

```python
op_mu, op_sigma, op_z_score = calculate_weighted_z_score(op_values, op_errors)
```

```python
op_chi_sq = np.sum(op_z_score**2, axis=1)
op_reduced_chi_sq = op_chi_sq / 6.
op_sector_chi_sq = np.sum(op_z_score**2, axis=0) / 240.
```

```python
plt.figure(figsize=(12,5))
plt.bar(np.arange(len(op_reduced_chi_sq)), op_reduced_chi_sq)
plt.grid(alpha=0.2)
```

```python
plt.figure(figsize=(12,5))
plt.bar(np.arange(len(op_sector_chi_sq)), op_sector_chi_sq)
plt.grid(alpha=0.2)
```

```python
def plot_compare_results(values1, errors1, values2, errors2, label1, label2):
    """ Function to compare data from two different distributions. """

    plt.figure(figsize=(8,6))
    plt.errorbar(np.arange(len(values1)), values1, errors1, label=label1, marker='o', linestyle='')
    plt.errorbar(np.arange(len(values2)), values2, errors2, label=label2, marker='o', linestyle='')
    plt.grid(alpha=0.2)
    plt.legend(frameon=False)
```

```python
plot_compare_results(values[12:24,1], errors[12:24,1], 
                     op_values[12:24,1], op_errors[12:24,1],
                    label1='Bootstrap', label2='One pass')
```

```python
plot_compare_results(values[12:24,3], errors[12:24,3], 
                     op_values[12:24,3], op_errors[12:24,3],
                    label1='Bootstrap', label2='One pass')
```

```python
plot_compare_results(op_values[12:24,1], op_errors[12:24,1], 
                     op_values[12:24,3], op_errors[12:24,3],
                    label1='One pass (S2)', label2='One pass (S4)')
```

```python
plot_compare_results(values[12:24,1], errors[12:24,1], 
                     values[12:24,3], errors[12:24,3],
                    label1='Bootstrap (S2)', label2='Bootstrap (S4)')
```

```python
def histogram_values_with_op_data(values_matrix, op_values, test_point):
    
    opts = {'edgecolor':'k', 'color':'blue', 'alpha':0.6,
           'bins':np.linspace(-0.2, 0.2, 40)}
    
    plt.figure(figsize=(16,10))

    for s in range(6):
        bias = np.mean(values_matrix[test_point, s, :] - op_values[test_point, s])
        
        plt.subplot(2,3,s+1)
        plt.hist(values_matrix[test_point, s, :] - op_values[test_point, s], **opts);
        plt.axvline(0.0, color='k')
        plt.grid(alpha=0.2)
        plt.title(f'Sector {s+1}, Bias = {bias}')
```

```python
histogram_values_with_op_data(values_matrix, op_values, test_point=5)
```
