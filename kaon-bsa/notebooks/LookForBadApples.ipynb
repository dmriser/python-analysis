```python
def get_values_and_errors(data_dict):
    """ From the data dictionary create
    an array of values and an array of errors
    with shape (npoints, nsectors). """
    
    npoints = len(data_dict[1])
    values = np.zeros(shape=(npoints,6))
    errors = np.zeros(shape=(npoints,6))
    
    for s in range(1,7):
        values[:,s - 1] = data_dict[s]['value'].values
        errors[:,s - 1] = data_dict[s]['stat'].values
    
    return values, errors
```

```python
def calculate_residuals(values, errors):
    """ Calculate point-by-point chi2. """
    
    npoints, nsectors = values.shape
    assert(values.shape == errors.shape)
    
    weights = 1.0 / errors**2
    mu = np.sum(values * weights, axis = 1) / np.sum(weights, axis = 1)
    sigma = 1 / np.sum(weights, axis = 1)
    mse = (values - np.multiply(np.ones(shape=(nsectors,npoints)), mu).transpose())**2
    chi2 = mse / np.multiply(np.ones(shape=(nsectors,npoints)), sigma).transpose()
    
    return mu, sigma, chi2

def calculate_z_scores(values, errors):
    """ Calculate point-by-point z-score. """
    
    npoints, nsectors = values.shape
    assert(values.shape == errors.shape)
    
    weights = 1.0 / errors**2
    mu = np.sum(values * weights, axis = 1) / np.sum(weights, axis = 1)
    sigma = np.sqrt(1 / np.sum(weights, axis = 1))
    residual = (values - np.multiply(np.ones(shape=(nsectors,npoints)), mu).transpose())
    z_score = residual / np.multiply(np.ones(shape=(nsectors,npoints)), sigma).transpose()
    
    return mu, sigma, z_score
```

```python
def plot_point_summary(values, errors, mu, sigma, index):
    
    plt.axvline(mu[index])
    plt.axvspan(mu[index] - sigma[index], mu[index] + sigma[index], alpha=0.4)
    
    for i,(v,e) in enumerate(zip(values[index,:], errors[index,:])):
        plt.errorbar(x=v, y=i, xerr=e, yerr=None, marker='o', color='k')
    
    plt.grid(alpha=0.2)
    plt.xlim([-0.1, 0.1])
    plt.yticks(np.arange(6),['Sector {}'.format(s+1) for s in range(6)])
```

```python
def weighted_mean(values, errors):
    weights = 1. / errors**2
    norm = np.sum(weights)
    return np.sum(values * weights / norm)
```

```python
def calculate_loocv(values, errors, test_point):
    
    cv_mu = np.zeros(6)
    cv_sig = np.zeros(6)
    
    for i in range(6):
        indices = [j for j in range(6) if j != i]
        cv_mu[i] = weighted_mean(values[test_point, indices],
                       errors[test_point, indices])
        cv_sig[i] = np.sqrt(1. / np.sum(1. / errors[test_point, indices]**2))
        
    return np.mean(cv_mu), np.std(cv_mu), np.mean(cv_sig), np.std(cv_sig)
```

```python
npoints = values.shape[0]
cv_mean = np.zeros(npoints)
cv_sig = np.zeros(npoints)

for i in range(npoints):
    cvmm, cvms, cvsm, cvss = calculate_loocv(values, errors, i)
    cv_mean[i] = cvmm
    cv_sig[i] = cvsm 
```

```python
def calculate_loocv_z_scores(values, errors):
    """ Calculate point-by-point z-score. """
    
    npoints, nsectors = values.shape
    assert(values.shape == errors.shape)
    

    mu = np.zeros(npoints)
    sigma = np.zeros(npoints)

    for i in range(npoints):
        cvmm, cvms, cvsm, cvss = calculate_loocv(values, errors, i)
        mu[i] = cvmm
        sigma[i] = cvsm 
    

    residual = (values - np.multiply(np.ones(shape=(nsectors,npoints)), mu).transpose())
    z_score = residual / np.multiply(np.ones(shape=(nsectors,npoints)), sigma).transpose()
    
    return mu, sigma, z_score
```
